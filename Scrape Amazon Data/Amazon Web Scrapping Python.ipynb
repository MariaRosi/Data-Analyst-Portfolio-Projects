```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Functions to extract the details
def get_title(soup):
    try:
         title = soup.find('span', attrs={'id':'productTitle'}).text.strip()
    except AttributeError:
        title = ""
    return title

def get_price(soup):
    try:
         price = soup.find('span', attrs={'class':'a-price aok-align-center reinventPricePriceToPayMargin priceToPay'}).text.strip()
    except AttributeError:
        price = ""
    return price


def get_rating(soup):
    try:
         rating = soup.find('i', attrs={'class':'a-icon a-icon-star a-star-4-5 cm-cr-review-stars-spacing-big'}).text.strip()
    except AttributeError:
        rating = ""
    return rating


def get_availability(soup):
    try:
         availability = soup.find('div', attrs={'id':'availability'}).text.strip()
    except AttributeError:
        availability = "Not Available"
    return availability



```

```python
# Webpage URL
URL = "https://www.amazon.co.uk/s?k=apple+watch&crid=1MJ989T72Z1IG&sprefix=apple+watch%2Caps%2C98&ref=nb_sb_noss_1"

# User Agent header
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

# HTTP Request
web_page = requests.get(URL, headers=HEADERS)

# Soup object by collecting all the data
web_soup = BeautifulSoup(web_page.content, 'html.parser')

# All products links
links_all_products = web_soup.find_all('a' , attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

# List for storing all the href links
list_all_products_href = []

# Dictionary for holding scrapping data
dict_scrapping_data = {'Title': [], 'Price': [], 'Rating': [], 'Availability': []}

for each_link in links_all_products:
    list_all_products_href.append("https://www.amazon.co.uk" + each_link.get('href'))


# Now looping the all product href list
for each_href in list_all_products_href:
    each_item_page = requests.get(each_href, headers=HEADERS)
    each_item_soup = BeautifulSoup(each_item_page.content, 'html.parser')

    #title = each_item_soup.find('span', attrs={'id':'productTitle'}).text.strip()
    dict_scrapping_data['Title'].append(get_title(each_item_soup))
    dict_scrapping_data['Price'].append(get_price(each_item_soup))
    dict_scrapping_data['Rating'].append(get_rating(each_item_soup))
    dict_scrapping_data['Availability'].append(get_availability(each_item_soup))


# DataFrame
df = pd.DataFrame.from_dict(dict_scrapping_data)

# Writing to csv file
df.to_csv('Amazon_WebScapping_Data.csv', header=True, index=False)

df




```
